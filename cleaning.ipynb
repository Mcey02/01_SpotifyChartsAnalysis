```python
## This cell ensures that all libraries are imported
import pandas as pd
```

```python
# This cell loads in the file. For now we are using only 1000 rows for testing and preparring cleaning. We only read in the columns that we will need
# dfile = pd.read_csv("../SpotifyData/spotify.csv", nrows=1000, usecols=['title','rank','date','artist','region','chart','trend','streams','popularity','af_danceability','af_energy','af_key','af_loudness','af_mode','af_speechiness','af_acousticness','af_instrumentalness','af_liveness','af_valence','af_tempo','af_time_signature'])

dfile = pd.read_csv("../SpotifyData/spotify.csv", usecols=['title','rank','date','artist','region','chart','trend','popularity','af_danceability','af_energy','af_key','af_loudness','af_mode','af_speechiness','af_acousticness','af_instrumentalness','af_liveness','af_valence','af_tempo','af_time_signature'])
print(dfile.head())
print(dfile.info())
```

```python
## Removal of these columns should result in about a 22% file reduction. --------------------  This has been moved to the read in step -------------------------
# dfile = dfile.drop(columns=["url", "track_id", "album", "duration_ms", "explicit", "available_markets", "release_date"])
# print(dfile.info())
```

```python
# These replacements will condense the file so as to save space and make calculations on trend easier
rfile = dfile.replace({"SAME_POSITION": 0, "MOVE_DOWN": -1, "MOVE_UP": 1, "NEW_ENTRY": 1, "top200": 2, "viral50": 5})
```

```python
## Checking types of transformed columns
print(rfile.info())
# print(rfile.head())
```

```python
# This cell will check if the nulls discovered below came before or after the save into the new csv file. CONCLUSION: The nulls appear before the new storage.
print(rfile.groupby("trend").count())
print(rfile.isnull().sum())
```

```python
# Saves edited df into new csv for easier future access
rfile.to_csv("../SpotifyData/dropped_transformed.csv", index=False)
```

```python
# Pulling in cleaned file from its stored version to make sure storage works properly
check_file = pd.read_csv("../SpotifyData/dropped_transformed.csv")
print(check_file.info())
print(check_file.head())
```

```python
# This checks and makes sure that the trend transformation correctly transforms the columns
print(check_file.groupby("trend").count())
```

```python
# The above cell revealed that some columns have null values that I didn't expect. Columns with no nulls are: rank date, region, and chart. 
# However, there seems to be a consistent number of data points that are null in the af and popularity columns. 
# This cell will serve to look at the entries that have null values in the af traits as these are what a lot of my analysis will be be based on
print(check_file.isnull().sum())
```

```python
# I am now doing analysis as to which rows have nulls and which don't
nulls = check_file[check_file["popularity"].isnull()]
print(nulls.head())
print(nulls.count())
```

```python
# This properly pulls entries without the null values in pop and af.
non_nulls = check_file[check_file["popularity"].notnull()]
print(non_nulls.head())
print(non_nulls.count())
```

```python
# We did not pull the songs missing the title or artist. I will see if I can manually fix these rows.
print(non_nulls.isnull().sum())
```

```python
# Pulls the null rows from the artist column for observance. It's one song
artist_nulls = non_nulls[non_nulls["artist"].isnull()]
print(artist_nulls.head(18))
```

```python
# Pulls the null rows from the title column for observance. It's one song
title_nulls = non_nulls[non_nulls["title"].isnull()]
print(title_nulls.head(11))
```

```python
# I don't think I can acurately fix these nulls so I am just going to remove them because it is only two songs. 
# I am going to concat all the null rows into one df and put them in their own file for storage.
all_nulls = pd.concat([title_nulls, artist_nulls, nulls])
```

```python
# This will pull the artist and title nulls from the working dataset. The correct number of rows have been pulled
non_t_nulls = non_nulls[non_nulls["title"].notnull()]
non_a_nulls = non_t_nulls[non_t_nulls["artist"].notnull()]
print(non_a_nulls.head())
print(non_a_nulls.count())
print(non_a_nulls.isnull().sum())
```

```python
## Finally I want to store the above two dataframes in files.
non_a_nulls.to_csv("../SpotifyData/final_data.csv", index=False)
all_nulls.to_csv("../SpotifyData/null_data.csv", index=False)
```
